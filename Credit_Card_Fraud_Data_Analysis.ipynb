```python
plt.figure(figsize=(20, 15))
for i, column in enumerate(credit_card.select_dtypes(include=['float64', 'int32']).columns, 1):
    plt.subplot(4, 4, i)
    credit_card.boxplot(column=[column])
    plt.title(column)
plt.show()
```

```python
numeric_columns = credit_card.select_dtypes(include=['float64','int32']).columns
for column in numeric_columns:
    Q1 = credit_card[column].quantile(0.25)
    Q3 = credit_card[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    credit_card[column] = credit_card[column].apply(lambda x: lower_bound if x < lower_bound else x)
    credit_card[column] = credit_card[column].apply(lambda x: upper_bound if x > upper_bound else x)
plt.figure(figsize=(20, 15))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(4, 4, i)
    credit_card.boxplot(column=[column])
    plt.title(column)
plt.show()
```
